```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
import re
```

```python
twee=pd.read_csv("C:/Users/Administrator/Desktop/Data visualization/BJP4Karnataka_tweets.csv")
twee.head(5)
```

```python
def wc(tweets,col):
    words=' '.join(tweets[col])
    colmask=np.array(Image.open("C:/Users/Administrator/Desktop/Data visualization/Karnataka.png"))
    STOPWORDS.update(['https','co'])
    stopwords=set(STOPWORDS) #removes reoccuring words such 'the','is'
    word_mod=re.sub("[^a-zA-Z0-9_]+"," ",words)
    wcd=WordCloud(background_color='black',mask=colmask,stopwords=stopwords,min_font_size=16).generate(word_mod)
    
    plt.figure(figsize=(20,10))
    plt.imshow(wcd)
    plt.tight_layout(pad=0)
    plt.axis("off")
    plt.show()
```

```python
wc(twee,'text')
```
