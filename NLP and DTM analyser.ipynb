### Co-occurance Analyser

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer 
import nltk
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_excel('C:/Users/Administrator/Desktop/Extra/feedback consultancy/R Analysis Test_excel data_2019_20.xlsx')
data.head()
```

```python
data['responses']=data['responses'].str.lower().str.replace('[^a-z ]','')
data.head()
```

```python
stopwords=nltk.corpus.stopwords.words('english')
stemmer=nltk.stem.LancasterStemmer()
```

```python
def clean_sentence(text):
    words=text.split(' ')
    words=[stemmer.stem(word) for word in words if word not in stopwords]
    return ' '.join(words)
def get_pos(text):
    tags=nltk.pos_tag(text)
    return tags
def tokenise(text):
    tokens=nltk.word_tokenize(text)
    words=[word for word in tokens if word not in stopwords]
    tag=nltk.pos_tag(words)
    return tag
```

```python
tagged=data['responses'].apply(tokenise)
```

```python
tagged_words=[]
eng=['NN','VBG','JJ']
for tags in tagged:
    for tag in tags:
        if tag[0]=='automation' or tag[1] in eng:
            tagged_words.append(tags)
        else: pass
```

```python
words=[]
for tags in tagged_words:
    for tag in tags:
        if tag[0]!='automation':
            words.append(tag[0])
        else: pass
```

```python
words # final words that are nouns, adjectives and verbs that occur with the word automation
```

### Sentiment Analyser

```python
data=pd.read_excel('C:/Users/Administrator/Desktop/Extra/feedback consultancy/R Analysis Test_excel data_2019_20.xlsx')
data=data.fillna(' ')
data.head()
```

```python
data['sentence']=data['Word 1'].astype(str)+' '+data['Word 2'].astype(str)+' '+data['Word 3'].astype(str)
data.head()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sentiment=SentimentIntensityAnalyzer()
```

```python
def get_sentiment(text):
    score=sentiment.polarity_scores(str(text))['compound']
    if score>0.5:
        return 'positive'
    elif score>=-0.5 and score<=0.5:
        return 'neutral'
    else:
        return 'negative'
    
```

```python
data['sentiment']=data['sentence'].apply(get_sentiment)
# data.drop('sentence')
data.head(10)
```

```python
docs=data['sentence'].str.lower().str.replace('[^a-z ]','')
```

```python
stopwords=nltk.corpus.stopwords.words('english')
stemmer=nltk.stem.PorterStemmer()
docs_clean=[]
def clean_sentence(text):
    words=text.split(' ')
    words_clean=[stemmer.stem(word) for word in words if word not in stopwords]
    return ' '.join(words_clean)
```

```python
docs_clean=docs.apply(clean_sentence)
```

```python
vectorizer=CountVectorizer()
vectorizer.fit(docs_clean)
dtm=vectorizer.transform(docs_clean)
dtm_df=pd.DataFrame(dtm.toarray(),columns=vectorizer.get_feature_names())
```

```python
dtm_df['sentiment']=data['sentiment']
```

```python
train_x,test_x=train_test_split(dtm_df,test_size=0.2,random_state=100)
train_y=data.iloc[train_x.index]['sentiment']
test_y=data.iloc[test_x.index]['sentiment']
```

```python
model=MultinomialNB()
model.fit(train_x,train_y)
pred=model.predict(test_x)
print(accuracy_score(pred,test_y))
```
